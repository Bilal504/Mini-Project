# 3) What is the most common Genre for Vin Diesel & Emma Watson movies?

```python
op1= data[["cast","genres"]]
op1
```

# The most common Genre for Emma Watson :

```python
op2=op1[op1["cast"].str.find("Emma Watson") != -1]
op3=op2.groupby("genres")["cast"].count()
op3
```

# The most common Genre for Vin diesel :

```python
op4=op1[ op1[ "cast" ].str.find("Vin Diesel")!= -1 ]
```

```python
op5= op4.groupby("genres")["cast"].count()
op5
```
