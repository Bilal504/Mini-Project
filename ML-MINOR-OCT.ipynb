# 3) What is the most common Genre for Vin Diesel & Emma Watson movies?

```python
def fil(a):
  #a=a.split('|')
  try:
    if 'Vin Diesel'in a :
      return True
    else:
      return False
  except:
    return False
```

```python
d=df.cast.str.split('|').apply(fil)
data3_vin=df[d]
len(df[d])
```

```python
temp_vin="|".join(data3_vin.genres)
temp_vin
```

```python
ans_3_a=Counter(temp_vin.split('|')).most_common(6)
ans_3_a=[*dict(ans_3_a)]
ans_3_a
```

```python
def fill(a):
  #a=a.split('|')
  try:
    if 'Emma Watson'in a :
      return True
    else:
      return False
  except:
    return False
```

```python
d2=df.cast.str.split('|').apply(fill)
data3_emma=df[d2]
len(data3_emma)
```

```python
data3_emma.head(3)
```

```python
temp_emma="|".join(data3_emma.genres)
temp_emma
```

```python
ans_3_b=Counter(temp_emma.split('|')).most_common(6)
ans_3_b=[*dict(ans_3_b)]
ans_3_b
```

```python
print("The Most common Genres between Emma Watson and Vin Diesel are :"," and  ".join(list(set(ans_3_a).intersection(set(ans_3_b)))))
```
